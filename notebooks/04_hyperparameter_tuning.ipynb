```python
import pandas as pd
import numpy as np
from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
import joblib
import warnings
warnings.filterwarnings('ignore')

# Load data
X_train = joblib.load('../models/X_train_scaled.joblib')
X_test = joblib.load('../models/X_test_scaled.joblib')
y_train = joblib.load('../models/y_train.joblib')
y_test = joblib.load('../models/y_test.joblib')

print("Data loaded successfully")
```

```python
# XGBoost hyperparameter tuning
print("=== XGBoost Hyperparameter Tuning ===")
xgb_params = {
    'max_depth': [5, 7, 9],
    'learning_rate': [0.05, 0.1, 0.2],
    'n_estimators': [50, 100]
}

xgb_grid = GridSearchCV(
    xgb.XGBClassifier(random_state=42, n_jobs=-1),
    xgb_params,
    cv=3,
    scoring='f1',
    n_jobs=-1,
    verbose=1
)
xgb_grid.fit(X_train, y_train)

print(f"Best params: {xgb_grid.best_params_}")
print(f"Best F1-Score (CV): {xgb_grid.best_score_:.4f}")

# Best model
best_xgb = xgb_grid.best_estimator_
```

```python
# Random Forest hyperparameter tuning (reduced)
print("\n=== Random Forest Hyperparameter Tuning ===")
rf_params = {
    'max_depth': [10, 15],
    'n_estimators': [50]
}

rf_grid = GridSearchCV(
    RandomForestClassifier(random_state=42, n_jobs=-1),
    rf_params,
    cv=2,
    scoring='f1',
    n_jobs=-1,
    verbose=1
)
rf_grid.fit(X_train, y_train)

print(f"Best params: {rf_grid.best_params_}")
print(f"Best F1-Score (CV): {rf_grid.best_score_:.4f}")

best_rf = rf_grid.best_estimator_
```
